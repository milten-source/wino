```python

import os
import urllib.request
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score



RANDOM_STATE = 42


DATA_PATH = "winequality-white.csv"
DATA_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"

if not os.path.exists(DATA_PATH):
    print("Nie znaleziono pliku lokalnego, pobieram dane...")
    urllib.request.urlretrieve(DATA_URL, DATA_PATH)

wine = pd.read_csv(DATA_PATH, sep=";")

print("Pierwsze 5 wierszy danych:")
display(wine.head())

print("Liczba wierszy i kolumn:")
print(wine.shape)

print("\nInformacje o danych:")
wine.info()

print("\nStatystyki opisowe:")
display(wine.describe().T)

print("\nBraki danych w kolumnach:")
display(wine.isna().sum())

print("\nLiczba zduplikowanych wierszy:")
print(wine.duplicated().sum())

print("\nRozklad zmiennej quality:")
display(wine["quality"].value_counts().sort_index())




quality_counts = wine["quality"].value_counts().sort_index()

plt.figure(figsize=(7, 4))
quality_counts.plot(kind="bar")
plt.title("Rozklad ocen jakosci wina")
plt.xlabel("Quality")
plt.ylabel("Liczba probek")
plt.show()


wine.hist(figsize=(14, 10), bins=30)
plt.suptitle("Histogramy cech", y=1.02)
plt.tight_layout()
plt.show()



plt.figure(figsize=(7, 4))
plt.scatter(wine["alcohol"], wine["quality"], alpha=0.25)
plt.title("Zaleznosc: alcohol vs quality")
plt.xlabel("Alcohol")
plt.ylabel("Quality")
plt.show()





wine_clean = wine.drop_duplicates().reset_index(drop=True)

print("Rozmiar przed czyszczeniem:")
print(wine.shape)

print("Rozmiar po usunieciu duplikatow:")
print(wine_clean.shape)

print("Usunieto wierszy:")
print(len(wine) - len(wine_clean))






X = wine_clean.drop(columns=["quality"])
y = wine_clean["quality"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE
)

print("X_train:")
print(X_train.shape)

print("X_test:")
print(X_test.shape)






models = {
    "Regresja liniowa": Pipeline([
        ("scaler", StandardScaler()),
        ("model", LinearRegression())
    ]),
    "Random Forest Regressor": RandomForestRegressor(
        n_estimators=200,
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
}

results = []
predictions = {}

for name, model in models.items():
    print("\nTrenowanie modelu:")
    print(name)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    predictions[name] = y_pred

    result = {
    "model": name,
    "RMSE": root_mean_squared_error(y_test, y_pred),
    "MAE": mean_absolute_error(y_test, y_pred),
    "R2": r2_score(y_test, y_pred)
}

    results.append(result)

results_df = pd.DataFrame(results).sort_values("RMSE")

print("\nWyniki modeli:")
display(results_df)






plt.figure(figsize=(7, 4))
plt.bar(results_df["model"], results_df["RMSE"])
plt.title("Porownanie modeli wedlug RMSE")
plt.ylabel("RMSE")
plt.xticks(rotation=20)
plt.show()





best_model_name = results_df.iloc[0]["model"]
best_predictions = predictions[best_model_name]

plt.figure(figsize=(6, 6))
plt.scatter(y_test, best_predictions, alpha=0.35)
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--"
)
plt.title("Wartosci rzeczywiste vs predykcje - " + best_model_name)
plt.xlabel("Rzeczywista jakosc")
plt.ylabel("Przewidywana jakosc")
plt.show()




rf_model = models["Random Forest Regressor"]

importances = pd.Series(
    rf_model.feature_importances_,
    index=X_train.columns
).sort_values(ascending=True)

plt.figure(figsize=(8, 5))
importances.plot(kind="barh")
plt.title("Waznosc cech - Random Forest")
plt.xlabel("Importance")
plt.show()

print("Najwazniejsze cechy wedlug Random Forest:")
display(importances.sort_values(ascending=False))




best_final_model = results_df.iloc[0]

print("\nWNIOSKI KONCOWE")
print("Celem projektu bylo przewidywanie jakosci bialego wina.")
print("Zmienna przewidywana byla kolumna quality, dlatego zastosowano regresje.")
print("Porownano modele: Regresja liniowa oraz Random Forest Regressor.")
print("Glowna metryka oceny byla RMSE. Im mniejsza wartosc RMSE, tym lepszy model.")
print("Najlepszy model:")
print(best_final_model["model"])
print("RMSE:")
print(best_final_model["RMSE"])





results_df.to_csv("wyniki_modeli.csv", index=False)

print("\nZapisano plik wyniki_modeli.csv")
```
